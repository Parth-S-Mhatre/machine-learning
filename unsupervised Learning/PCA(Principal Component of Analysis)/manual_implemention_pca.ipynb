```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_X=scaler.fit_transform(df)
```

```python
scaled_X-scaled_X.mean(axis=0)
```

```python
scaled_X
```

```python
covariance_matrix=np.cov(scaled_X,rowvar=False)
```

```python
eigen_values,eigen_vector=np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vector.shape
```

```python
# PCA 30 features Space --> 30 Principal Space -->2
# how many PC to choose ? N=2 --> top two Eigen values
```

```python
num_components=2
```

```python
#[0,1,2]
np.argsort([2,1,3])
```

```python
np.argsort(eigen_values)[::-1][:num_components]
```

```python
sorted_key=np.argsort(eigen_values)[::-1][:num_components] # reverse the order from higher to lower and then top two values
```

```python
eigen_values,eigen_vector=eigen_values[sorted_key],eigen_vector[:,sorted_key]
```

```python
#Original data -->project -->eigen vector linear Transformation
principal_component=np.dot(scaled_X,eigen_vector)
```
