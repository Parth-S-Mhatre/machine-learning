```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# .jpg .png-> numpy array
# this will covert the Image of jpg png into Numpy array
```

```python
image_as_array=mpimg.imread("images.jpeg")
```

```python
image_as_array.shape # (Height,width,color Channel)
```

```python
#plt.figure(dpi=200)
plt.imshow(image_as_array)
```

```python
#(H,W,C)-->2d (H*W,C)
```

```python
(h,w,c)=image_as_array.shape
```

```python
h
```

```python
w
```

```python
c
```

```python
image_as_array2d=image_as_array.reshape(h*w,c)
```

```python
len(image_as_array2d.shape) #2d
```

```python
len(image_as_array.shape) #3d
```

```python
from sklearn.cluster import  KMeans
```

```python
model=KMeans(n_clusters=6)
```

```python
labels=model.fit_predict(image_as_array2d)
```

```python
182*276 # data points
```

```python
labels
```

```python
model.cluster_centers_
```

```python
rgb_codes=model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes[labels]
```

```python
quantized_img=np.reshape(rgb_codes[labels],(h,w,c))
```

```python
quantized_img
```

```python
plt.imshow(quantized_img)
```
