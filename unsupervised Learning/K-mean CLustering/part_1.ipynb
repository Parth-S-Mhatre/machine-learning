```python
X=pd.get_dummies(df)
```

```python
from sklearn.preprocessing import  StandardScaler
```

```python
scaler=StandardScaler()
```

```python
Scaled_X=scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
help(KMeans)
```

```python
model=KMeans(n_clusters=2)
```

```python
#fit_transform 
cluster_label=model.fit_predict(Scaled_X)
#fit means it perform the clustering algo on the data like putting value of the n_cluster
#predict means it find the Closest cluster for each data point 
```

```python
cluster_label
```

```python
X['Cluster']=cluster_label
```

```python
plt.figure(figsize=(12,8),dpi=200)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
plt.xticks(rotation=90);
```

```python
ssd=[] # sum of sqaure disatnce between each data points 
for k in range(2,10):
    model=KMeans(n_clusters=k)
    model.fit(Scaled_X)
    ssd.append(model.inertia_) #ssd-cluster center
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd,'o--')
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff() #this diff is the difference between it's current value and the previous one 
```
