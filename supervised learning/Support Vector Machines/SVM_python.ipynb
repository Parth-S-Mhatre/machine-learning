```python
from sklearn.svm import SVC
```

```python
y=df['Virus Present']
```

```python
X=df.drop('Virus Present',axis=1)
```

```python
model=SVC(kernel='linear',C=1000)
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import  plot_svm_boundary
```

```python
plot_svm_boundary(model,X,y)
```

```python
model=SVC(kernel='linear',C=0.005) # c is inverse proportional to the margin 
model.fit(X,y)
from svm_margin_plot import  plot_svm_boundary
plot_svm_boundary(model,X,y)
```

```python
model=SVC(kernel='rbf',C=1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model=SVC(kernel='rbf',C=0.1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
help(SVC)
```

```python
model=SVC(kernel='rbf',C=0.1,gamma=2) # as gamma increase the kernel will be more dense i.e it is covering all the support vectors
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model=SVC(kernel='rbf',C=0.1,gamma='scale') # by default
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model=SVC(kernel='sigmoid',C=1) # by default
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model=SVC(kernel='poly',C=0.05,degree=4) # by default
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm=SVC()
param_grids={'C':[0.1,0.01,1],'kernel':['linear','rdf']}
```

```python
grid=GridSearchCV(svm,param_grids)
grid.fit(X,y)
```
