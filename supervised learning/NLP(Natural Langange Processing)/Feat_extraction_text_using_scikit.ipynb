```python
text=['This is  a line',
      'This is another line',
      'Completely Different line']
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
```

```python
help(CountVectorizer)
```

```python
cv=CountVectorizer(stop_words='english')  #remove the common word like the a u can also pass custom stop word i.e common word...etc 
```

```python
sparse_matrix=cv.fit_transform(text)
```

```python
sparse_matrix.todense()
1
```

```python
cv.vocabulary_  # unique word
```

```python
tfdf=TfidfTransformer()
```

```python
sparse_matrix
```

```python
results=tfdf.fit_transform(sparse_matrix)
```

```python
results.todense()
```
