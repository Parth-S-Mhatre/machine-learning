```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
df=pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\parth\\Machine-learning\\Machine-learning\\08-Linear-Regression-Models\\Advertising.csv")
```

```python
df.head()
```

```python
# linear reg mein multiple feature ko handle karna difficult hota hai so advertising aur sales ka relation nikale ne ke liye we have to sum total spend 
# on advertising 
df['total spend']=df['TV']+df['newspaper']+df['radio']
```

```python
df.head()
```

```python
sns.scatterplot(data=df,x='total spend',y='sales')
```

```python
# to draw line of linear regression
sns.regplot(data=df,x='total spend',y='sales')
```

```python
X=df['total spend']
y=df['sales']
```

```python
# y=mx+b
#y=b1x+b0
#help(np.polyfit) Least squares polynomial fit
```

```python
np.polyfit(x=X,y=y,deg=1) # this will give b1 and b0 
# we can do prediction 
```

```python
potential_spend=np.linspace(0,500,100)
```

```python
#y=b1x+b0
predicted_sales=0.04868788*potential_spend+4.24302822
```

```python
sns.scatterplot(data=df,x='total spend',y='sales')## y=mx+b
plt.plot(potential_spend,predicted_sales,color='red') # this what replot was giving to us  
```

```python
# for example if i want to spend 200 to ad how much sale will earn 
spend=200
#y=b1x+b0 only change X 
predicted_sales=0.04868788*spend+4.24302822

predicted_sales # now see in the graph whether it is right or wrong 

```

```python
# higher order polynomial regression 
np.polyfit(X,y,deg=3)
# y=b1x+bo


# y=b3*x**3+b2*x**2+b1*x+b0
```

```python
pot_sales=np.linspace(0,500,100)
predict_sales=3.07615033e-07*pot_sales**3+ -1.89392449e-04*pot_sales**2+8.20886302e-02*pot_sales+2.70495053e+00

```

```python
plt.figure(dpi=200)
sns.scatterplot(data=df,x='total spend',y='sales')## y=mx+b
plt.plot(pot_sales,predict_sales,color='red')
```
