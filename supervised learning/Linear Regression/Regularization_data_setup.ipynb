```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\parth\\Machine-learning\\Machine-learning\\08-Linear-Regression-Models\\Advertising.csv")
```

```python
X=df.drop('sales',axis=1)
```

```python
y=df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter=PolynomialFeatures(degree=3,include_bias=False)
```

```python
poly_features=polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(X)
```

```python
X_train=scalar.fit_transform(X_train)
```

```python
X_test=scalar.fit_transform(X_test)
```

```python
X_train[0]
```

Ridge Regression 

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge(alpha=10)
```

```python
ridge.fit(X_train,y_train)
```

```python
test_prediction=ridge.predict(X_test)
```

```python
test_prediction
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE=mean_absolute_error(y_test,test_prediction)
```

```python
MSE=mean_squared_error(y_test,test_prediction)
```

```python
RSME=np.sqrt(MSE)
```

```python
MAE
```

```python
RSME
```

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge_cv_model=RidgeCV(alphas=(0.1,1.0,10.0),scoring='neg_mean_absolute_error') # higher the negative mean absolute better it is because it is negative 
```

```python
ridge_cv_model.fit(X_train,y_train)
```

```python
ridge_cv_model.alpha_ # this tell which alpha parameter performance best 
```

```python
#from sklearn.metrics import SCORERS
```

```python
test_prediction=ridge_cv_model.predict(X_test)
```

```python
MAE=mean_absolute_error(y_test,test_prediction)
```

```python
MSE=mean_squared_error(y_test,test_prediction)
```

```python
RSME=np.sqrt(MSE)
```

```python
MAE
```

```python
RSME
```

```python
ridge_cv_model.coef_
```

# LASSO 
Least absolute shrinkage and selection operator

```python
from sklearn.linear_model import LassoCV
```

```python
lassocv_model=LassoCV(eps=0.001,n_alphas=100,cv=10,max_iter=10000000) # eps=min_aplha/max_alpha N_alhas=Number of times alpha u want cv number of folds in cross validation 
```

```python
lassocv_model.fit(X_train,y_train)
```

```python
lassocv_model.alpha_
```

```python
test_prediction=lassocv_model.predict(X_test)
```

```python
MAE=mean_absolute_error(y_test,test_prediction)
```

```python
MSE=mean_squared_error(y_test,test_prediction)
RSME=np.sqrt(MSE)
```

```python
MAE
```

```python
RSME
```

```python
lassocv_model.coef_
```
