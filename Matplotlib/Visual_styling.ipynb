```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,11)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
# Plot India's GDP
axes.plot(x, x, label="India's GDP",color='#04336b',linewidth=1)    # accept Hex code or RGB

axes.plot(x, x**4, label="India's GST",color='#9514E6',linewidth=1) 

axes.legend(loc=(0.025,0.85))
```

Line styling

```python
# possible linestype options --, , -., :, steps
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
# Plot India's GDP
axes.plot(x, x, label="India's GDP",color='#04336b',linewidth=1,ls='--')    # accept Hex code or RGB

axes.plot(x, x**4, label="India's GST",color='#9514E6',linewidth=1,linestyle='-.') 

axes.legend(loc=(0.025,0.85))
```

Custom Dash line

```python

fig=plt.figure()
axes=fig.add_axes([0,0,1,1])

line=axes.plot(x, x**4, label="India's GST",color='#9514E6',linewidth=1) 
line[0].set_dashes([2,5,2,5]) #  formula list[(point kitne two),(blank kitne space),(point space),(blank space)]

axes.legend(loc=(0.025,0.85))
```

Marker means line ke adhar points show karna basically relationship

```python
# possible linestype options --, , -., :, steps
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
# Plot India's GDP
axes.plot(x, x, label="India's GDP",color='#04336b',linewidth=1,ls='--',marker='o',ms=5)    # accept Hex code or RGB
# can also google search for different marker 
axes.plot(x, x**4, label="India's GST",color='#04336b',linewidth=1,ls='--',marker='+',markersize=5) 
axes.legend(loc=(0.025,0.85))
```

Marker editing

```python
# possible linestype options --, , -., :, steps
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
# Plot India's GDP
# Plot India's GDP with specified styling
axes.plot(x, x, label="India's GDP", color='#04336b', linewidth=1, ls='--', marker='s', ms=5,
           markerfacecolor='green', markeredgewidth=4, markeredgecolor='orange')
```
