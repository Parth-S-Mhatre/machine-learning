```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
a=np.linspace(0,10,11)
```

```python
b=a**4
```

```python
fig=plt.figure(figsize=(2,2),dpi=200) # figure size in inches and dpi means dots per inch something like resolution 
axes=fig.add_axes([0,0,1,1])
axes.plot(a,b)
fig.savefig('New_figure',bbox_inches='tight') # bbox_inches save the axes the name 
```
