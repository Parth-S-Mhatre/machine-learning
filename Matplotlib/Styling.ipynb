```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,11)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
# Plot India's GDP
axes.plot(x, x, label="India's GDP")   

# Plot India's GST
axes.plot(x, x**2, label="India's GST") 
# calling the legend function
# Add a legend
axes.legend(loc=(0.050,0.8))  # x,y coordinates to place the legend

plt.show()
```
