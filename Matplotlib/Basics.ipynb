```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x=np.arange(0,10)
```

```python
y=2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) # the graph will be x vs y 
#plt.show() # this is used to display the graph for py script
plt.xlabel('X axis') # to add the title to the x axis 
plt.ylabel('Y axis')# # to add the title to the y axis 
plt.xlim(0,8) # to set the limit of the x axis 
plt.ylim(0,18)# to set the limit of the y axis 
plt.title('String title') # to add the title of the graph 
plt.savefig('myfirstplot.png') # to save the graph as a file like image -: png,jpg etc
```
