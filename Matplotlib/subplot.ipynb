```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.linspace(0,10,11)
```

```python
b=a**4
```

```python
x=np.arange(0,10)
```

```python
y=2*x
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2) # just like matrix  and axes is numpy array 
axes[0].plot(x,y)     # when row is =1 and columns is 2 just like array 
axes[1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2) # just like matrix  and axes is numpy array 
axes[0][0].plot(x,y)     
# axes[1].plot(a,b) as the subplot is no longer 1D array its now 2d 
axes[1][1].plot(a,b)
```

Spacing and arrangement 

```python
fig,axes=plt.subplots(nrows=3,ncols=1) 
axes[0].plot(x,y)
plt.tight_layout() # gives the spacing between indivdual graph     
#axes[1].plot(a,b)
```

Manual Spacing

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(4,10),dpi=200) 
# axes 00
axes[0][0].plot(x,y)
axes[0][0].set_title('Title of 00')
axes[0][0].set_xlabel('X axis')
axes[0][0].set_ylabel('Y axis')
axes[0][0].set_xlim(2,6)

# axes 11
axes[0][1].plot(x,y)
axes[0][1].set_title('Title of 01')
axes[0][1].set_xlabel('X axis')
axes[0][1].set_ylabel('Y axis')

axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
# to give main title 
fig.suptitle("MAIN title",fontsize=16) # fontsize in px 
fig.subplots_adjust(wspace=1,hspace=0.5)
fig.savefig("new_subplot.png",bbox_inches='tight')
```
