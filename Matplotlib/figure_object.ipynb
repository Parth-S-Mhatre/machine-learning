Multiple Axes how to plot it 

```python
fig=plt.figure()
# Larger axes
axes1=fig.add_axes([0,0,1,1]) # first two parameter are to set the origin of the graph,last two parameter are the size of the graph means width and height 
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# Smaller axes
axes2=fig.add_axes([0.2,0.5,0.25,0.25]) # first two parameter are set from 20% from both axes (x,y) and last two parameter are the half the size of the figure
axes2.plot(x,y)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```
