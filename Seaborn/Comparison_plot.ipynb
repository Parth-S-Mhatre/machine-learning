```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\parth\\Machine-learning\\Machine-learning\\05-Seaborn\\StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.jointplot(data=df,x='math score',y='reading score')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='hex')  # like for more clearity in the graph or plot
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='hist') # historgram 
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',fill=True) # kde 
```

```python
sns.jointplot(data=df,x='math score',y='reading score',hue='gender')
```

Pair plot it compare only data which has numeric value 

```python
sns.pairplot(data=df,hue='gender') # as u add another columns for comparison it always convert to kde because in histogram it overide the shades
```

```python
sns.pairplot(data=df,hue='gender',diag_kind='hist')
```
