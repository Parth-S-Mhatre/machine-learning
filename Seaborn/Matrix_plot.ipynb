```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\parth\\Machine-learning\\Machine-learning\\05-Seaborn\\country_table.csv")
```

```python
df.head()
```

```python
df=df.set_index("Countries")
```

```python
df
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.heatmap(df.drop("Life expectancy",axis=1),linewidths=0.5,annot=True) # decide to drop the life expectancy column because uski scale alag ban rahi hai 63 se 80 which is diferrent
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.heatmap(df) 
```

Color Mapping 

```python
plt.figure(figsize=(10,5),dpi=200)
sns.heatmap(df.drop("Life expectancy",axis=1),linewidths=0.5,annot=True,cmap='viridis',cbar=bool)
```

Cluster map do's grouping of similar columns for Example like America and Europe are developed so they can be group together 

```python
plt.figure(figsize=(10,5),dpi=200)
sns.clustermap(df.drop("Life expectancy",axis=1),linewidths=0.5,annot=True)
```

```python
# if u want to grouping only index 
plt.figure(figsize=(10,5),dpi=200)
sns.clustermap(df.drop("Life expectancy",axis=1),linewidths=0.5,annot=True,col_cluster=False)
```
