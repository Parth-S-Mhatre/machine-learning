```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\parth\\Machine-learning\\Machine-learning\\05-Seaborn\\StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,x='parental level of education',y='math score',hue='test preparation course')
plt.legend(bbox_to_anchor=(-0.1,0.5))
```

Violin plot 

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',hue='test preparation course',split=True)
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',hue='test preparation course',split=True,inner=None) # inner middle line ko nikal tha 
# hai aur split do violinplot join karke ban ek tha hai 
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',split=True,inner='quartile') # inner middle line ko nikal tha 
# hai aur split do violinplot join karke ban tha hai 
```

Swarm plot

```python
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course') # it is difficult to read who completed or none 
```

```python
# solution for it 
plt.figure(figsize=(10,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxenplot(data=df,x='math score',y='gender',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,x='math score',y='gender',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.5))
```
