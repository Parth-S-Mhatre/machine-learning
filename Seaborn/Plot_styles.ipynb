```python
# historgrams show distribution 
# to set the styling of grid
sns.set(style='darkgrid')
plt.figure(figsize=(10,8),dpi=200)
sns.displot(data=df,x='salary',bins=20,color='red',edgecolor='blue',linewidth=4,ls='--') # this all agruments to edit the historgram
```

```python
sns.displot(data=df,x='salary',bins=20,color='red')
```

```python
sns.displot(data=df,x='salary',bins=20,color='red',kde=True,rug=True) # dist plot can show all the types of plot rug,historgram,kenel density
```

```python
sns.histplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_age=np.random.randint(0,100,200)
```

```python
sample_age
```

```python
sample_age=pd.DataFrame(sample_age,columns=['Age'])
```

```python
sample_age
```

```python
sns.displot(data=sample_age,x='Age',bins=30,rug=True,kde=True)
```

```python
sns.kdeplot(data=sample_age,x='Age')  # it's going to negative also like but in reality age cannot be negative but still it's showing probability of it 
```

```python
# to show like problme of this 
sns.kdeplot(data=sample_age,x="Age",clip=(0,100)) # clip matlab guess karo like real life mein age 0 se 100 tak ja sakti hai 
```

```python
# now it's correct 
sns.kdeplot(data=sample_age,x="Age",clip=(0,100),bw_adjust=0.5,shade=True) # if u increase bandwidth it will like normal distribution which is not useful 
```
