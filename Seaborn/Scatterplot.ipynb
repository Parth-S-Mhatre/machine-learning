```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns 
```

```python
df=pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\parth\\Machine-learning\\Machine-learning\\05-Seaborn\\dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=200)     # matlablib and seaborn are inter-connected to each other 
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education') # hue is basically it color or differentiate like for catergory columns it will show different
# for continous same color but just variation in color shades
```

Continous Columns

```python
plt.figure(figsize=(12,4),dpi=200)     # matlablib and seaborn are inter-connected to each other 
sns.scatterplot(x='salary',y='sales',data=df,hue='salary')
```

To Change the color of points

```python
plt.figure(figsize=(12,4),dpi=200)     # matlablib and seaborn are inter-connected to each other 
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2') # Go to the documentation of matplotlib
```

to differentiate based on salary like jinki jada salary unka bade points 

```python
plt.figure(figsize=(12,4),dpi=200)    
sns.scatterplot(x='salary',y='sales',data=df,size='salary')
```

To increase size in general 

```python
plt.figure(figsize=(12,4),dpi=200)    
sns.scatterplot(x='salary',y='sales',data=df,s=200,alpha=0.2) # alpha is basically transparency  s is normal size 
```

Styling 

```python
plt.figure(figsize=(12,4),dpi=200)    
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education') # style is apply on catergorical columns only 
plt.savefig("mySeaborn_plot.jpg")
```
