Statistical Estimation within Categories

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\parth\\Machine-learning\\Machine-learning\\05-Seaborn\\dm_office_sales.csv")
```

```python
df['division'].value_counts() # non visualization 
```

```python
plt.figure(figsize=(10,5),dpi=200) #visualization 
sns.countplot(data=df,x='division',hue='division') # we do not need to specifies y axis because it always be count 
```

```python
plt.figure(figsize=(10,5),dpi=200) #visualization 
sns.countplot(data=df,x='level of education',hue='division',palette='Set2')
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd',hue='division')  # to calculate level education per salary matlab jese jese education 
# badenga wese wese salary bhi badengi also ci=sd means standard deviation aur estimator =np.any numpy function 
# to place legend at any specific location
plt.legend(bbox_to_anchor=(1.05,1))
```
